```python
# important python libraries for data analysis and visualization

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings as wr
wr.filterwarnings('ignore')
```

```python
# pandas for reading data from various file formats (like CSV, Excel), manipulating and cleaning data

df = pd.read_csv('banknifty_data.csv')
```

```python
df.columns # command prints the columns in dataframe
```

```python
# Convert 'Date' column to datetime with the original format for time series data and analysis

df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
# Change the format to display only the year

df['Year'] = df['Date'].dt.strftime('%Y')
```

```python
# command sets the 'Date' column as the index of the dataFrame df and modifies the dataFrame in place without creating a new dataframe.

df.set_index('Date', inplace=True)
```

### Data Visualization:
## Visualize the data to identify patterns, trends, and outliers.

```python
# Plot the 'nifty bank' index over time to see its trend.

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['nifty bank'], label='nifty bank')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Nifty Bank Index Over Time')
plt.legend()
plt.show()
```

```python
# we can plot specific banks over time

plt.figure(figsize=(10, 6))
plt.plot(df.index, df['hdfc_bank'], label='HDFC BANK')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('HDFC bank Over Time')
plt.legend()
plt.show()
```

```python
# Box plots for each bank to understand the distribution and identify outliers.

plt.figure(figsize=(12, 8))
df.boxplot(column=['axis_bank', 'bandhan_bank', 'au_bank', 'sbi_bank', 'bob_bank', 'idfc_bank', 'federal_bank', 'kotak_bank', 'hdfc_bank', 'icici_bank', 'pnb_bank', 'indusind_bank'])
plt.title('Box Plot of Bank Stock Prices')
plt.xticks(rotation=45)
plt.show()
```

```python
# creates a bar plot to visualize the comparison of these mean stock prices BETWEEN different banks.

mean_prices = df.drop(columns=['nifty bank']).mean()


plt.figure(figsize=(12, 6))
mean_prices.plot(kind='bar', color='grey')
plt.xlabel('Banks')
plt.ylabel('Mean Stock Price')
plt.title('Comparison of Mean Stock Prices of Component Banks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
corr_matrix = df.drop(columns=['nifty bank']).corr()

# Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='Pastel2')
plt.title('Correlation Matrix of Bank Stock Prices')
plt.show()
```

```python
# Calculation of 50-day and 200-day moving averages for the Bank Nifty index

df['nifty bank_50d'] = df['nifty bank'].rolling(window=50).mean()
df['nifty bank_200d'] = df['nifty bank'].rolling(window=200).mean()

# Plot the moving averages

plt.figure(figsize=(14, 7))
plt.plot(df.index, df['nifty bank'], label='Bank Nifty Index', color='blue')
plt.plot(df.index, df['nifty bank_50d'], label='50-Day Moving Average', color='orange')
plt.plot(df.index, df['nifty bank_200d'], label='200-Day Moving Average', color='green')
plt.xlabel('Date')
plt.ylabel('Index Value')
plt.title('Bank Nifty Index with 50-Day and 200-Day Moving Averages')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# we can also plot specific banks moving averages
# Calculation of 50-day and 200-day moving averages for the icici bank index

df['icici_bank_50d'] = df['icici_bank'].rolling(window=50).mean()
df['icici_bank_200d'] = df['icici_bank'].rolling(window=200).mean()

# Plot the moving averages

plt.figure(figsize=(14, 7))
plt.plot(df.index, df['icici_bank'], label='Bank Nifty Index', color='blue')
plt.plot(df.index, df['icici_bank_50d'], label='50-Day Moving Average', color='red')
plt.plot(df.index, df['icici_bank_200d'], label='200-Day Moving Average', color='brown')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('ICICI Bank with 50-Day and 200-Day Moving Averages')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
